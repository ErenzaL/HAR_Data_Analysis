```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
data_path = os.path.join(os.getcwd(), 'data','A_DeviceMotion_data')
if os.path.exists(data_path):
    print(f"데이터 경로 확인: {data_path}")
else:
    print("경로를 찾을 수 없습니다.")
```

```python
# 모든 데이터를 저장할 리스트
all_data = []

# 폴더 전체 탐색
for root, dirs, files in os.walk(data_path):
    for file in files:
        if file.endswith(".csv"):
            file_path = os.path.join(root, file)

            # CSV 읽기
            df = pd.read_csv(file_path, index_col=0)

            # 파일 이름 정보 추가
            df["source_file"] = file

            # 리스트에 저장
            all_data.append(df)

# 모든 데이터 하나로 합치기
final_df = pd.concat(all_data, ignore_index=True)
```

```python
# 결과 확인
print("데이터 크기:", final_df.shape)
display(final_df.head())

print("데이터 통계:")
display(final_df.describe()) #기초 통계 + 이상치 알기 위해
```

```python
def calculate_magnitude(df, x_col, y_col, z_col):
    """
    x, y, z 컬럼을 받아 벡터의 크기(Magnitude)를 계산하는 함수 #힘의 크기 알게 위해
    """

    x = df[x_col]
    y = df[y_col]
    z = df[z_col]

    # magnitude = sqrt(x^2 + y^2 + z^2)
    magnitude = np.sqrt(x**2 + y**2 + z**2)

    return magnitude
```

```python
# 함수 적용 및 새로운 컬럼 생성
final_df['mag_acc'] = calculate_magnitude(
    final_df,
    'userAcceleration.x',
    'userAcceleration.y',
    'userAcceleration.z'
)

# 결과 확인
print(final_df[['userAcceleration.x', 'mag_acc']].head())
```

```python
BASE_PATH = data_path

csv_path = f"{BASE_PATH}/processed_sensor_data.csv"
pkl_path = f"{BASE_PATH}/sensor_data.pkl"

final_df.to_csv(csv_path, index=False)
final_df.to_pickle(pkl_path)

print("저장 완료")
```
